Descripción de variables
1. CRIM: Tasa de criminalidad per cápita por ciudad.

cuadrados.
3. INDUS: Proporción de acres comerciales no minoristas por ciudad.
4. CHAS: Variable ficticia de Charles River (1 si el tramo limita con el río, 0 en caso contrario).
5. NOX: Concentración de óxidos nítricos (partes por 10 millones).
6. RM: Número promedio de habitaciones por vivienda.
7. AGE: Proporción de unidades ocupadas por sus propietarios construidas antes de 1940.
8. DIS: Distancias ponderadas a cinco centros de empleo de Boston.
9. RAD: Indice de accesibilidad a carreteras radiales.
10. TAX: Tasa de impuesto a la propiedad de valor total por \$10,000.
11. PTRATIO: Proporción alumno-profesor por localidad.
12. B: 1000\times (Bk-0.63)2
, donde Bk es la proporción de personas de ascendencia
afroamericana por ciudad.
13. LSTAT: Porcentaje de población de menor status.
14. MEDV: Valor medio de las viviendas ocupadas por sus propietarios en miles de dólares.

gráfico de densidad , asimetría y curtosis de 3 variables:

---



```r
densidad = density(data$medv)
plot(densidad, main = "medv", col = "orange", lwd = 3, xlab = "Valor medio")

library(moments)
skewness(data$medv)#simetría
kurtosis(data$medv)#curtosis
```

 Simetria: asimetría positiva (cola larga a la derecha), muestra como la mayoría de los valores de vivienda están en el rango medio/bajo, pero hay algunas viviendas de alto valor que elevan la media

 Curtosis: leptocúrtica (más puntiaguda que la normal) alta sugiere que la distribución tiene un pico más pronunciado, con más valores cerca de la media y colas más largas de lo normal.

Conclusion: estoi deacuerdo, mientras menos habitaciones menor el precio.

```r
densidad = density(data$lstat)
plot(densidad, main = "lstat", col = "orange", lwd = 3, xlab = "% de bajos ingresos")

library(moments)
skewness(data$lstat)#simetría
kurtosis(data$lstat)#curtosis
```

Simetria: asimetría positiva (cola larga a la derecha), muestra que la mayoría de los barrios tienen una baja proporción de personas con bajos ingresos con algunas exepciones.

Curtosis: leptocúrtica, la distribución es moderadamente puntiaguda, pero no muy diferente de la normal.

Conclusion: mas o menos lo mismo que la anterior ya que estan relacionados.

```r
densidad = density(data$rm)
plot(densidad, main = "rm", col = "orange", lwd = 3, xlab = "Promedio de habitaciones")

library(moments)
skewness(data$rm)#simetría
kurtosis(data$rm)#curtosis
```

Simetria: casi simétrica(cercana a 0), con una ligera inclinación hacia valores más altos (asimetría positiva leve)

Curtosis: la mayoría de los valores están concentrados alrededor del promedio y que existen algunas viviendas con muchas más habitaciones, aunque son menos comunes.

Conclusion: concuerdo ya que la mayoria de las casas tienen en promedio el mismo tamaño.

Grafico de barras De variables relacionadas de chas:

---



```r
par(mfrow = c(1, 2))

medv_prom = tapply(data$medv, data$chas, mean)

crim_prom = tapply(data$crim, data$chas, mean)

barplot(medv_prom,
        names.arg = c("No río", "Cerca del río"),
        col = "lightblue",
        main = "Valor medio por cercanía al río",
        ylab = "medv")

barplot(crim_prom,
        names.arg = c("No río", "Cerca del río"),
        col = "lightgreen",
        main = "criminalidad promedio por cercanía al río",
        ylab = "crim")
```

en medv el valor de las casas cerca de del rio aumenta y en cuanto a crim estando lejos del rio al disminuir el valor de las casas el nivel socioeconomico disminuye aumentando la criminalidad.
